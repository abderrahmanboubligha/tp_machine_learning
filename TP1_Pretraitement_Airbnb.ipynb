```python
#categorical variable
data["cancellation_policy"] = data["cancellation_policy"].fillna(data["cancellation_policy"].mode()[0])
```

```python
data["Construction year"] = data["Construction year"].fillna(data["Construction year"].mean())
data["Construction year"]= data["Construction year"].astype(int)
data
```

```python
#mean when the destribution is normal and median when the distribution is deferent some so big and some so small
data["price"] = (
    data["price"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace(" ", "", regex=False)
)
data["price"] = data["price"].astype(float)
data
```

```python
# Get summary stats for non-null values
stats = data['price'].describe()
skew = data['price'].skew()

print(f"Mean: {stats['mean']}, Median: {stats['50%']}, Skew: {skew}")

# Decision logic
if abs(skew) < 0.5:
    print("Strategy: Mean (Symmetric)")
else:
    print("Strategy: Median (Skewed)")
```

```python
data["price"] = data["price"].fillna(data["price"].mean())
data["service fee"] = (
    data["service fee"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace(" ", "", regex=False)
)
data["service fee"] = data["service fee"].astype(float)

# Get summary stats for non-null values
stats = data['service fee'].describe()
skew = data['service fee'].skew()

print(f"Mean: {stats['mean']}, Median: {stats['50%']}, Skew: {skew}")

# Decision logic
if abs(skew) < 0.5:
    print("Strategy: Mean (Symmetric)")
else:
    print("Strategy: Median (Skewed)")
```
