```python
# Get summary stats for non-null values
stats = data['price'].describe()
skew = data['price'].skew()

print(f"Mean: {stats['mean']}, Median: {stats['50%']}, Skew: {skew}")

# Decision logic
if abs(skew) < 0.5:
    print("Strategy: Mean (Symmetric)")
else:
    print("Strategy: Median (Skewed)")
```

# Gestion des valeurs aberrantes

```python
columns = ["price", "minimum nights", "number of reviews"]

for col in columns:
    plt.figure()
    sns.boxplot(x=data[col])
    plt.title(f"Boxplot de {col}")
    plt.show()
```

```python
def remove_outliers_all(df, cols):
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        df = df[(df[col] >= lower) & (df[col] <= upper)]
    
    return df

data = remove_outliers_all(data, ["price", "minimum nights", "number of reviews"])
```

```python
columns = ["price", "minimum nights", "number of reviews"]

for col in columns:
    plt.figure()
    sns.boxplot(x=data[col])
    plt.title(f"Boxplot de {col}")
    plt.show()
```

# Transformation des Données

```python
data
```

```python
# Nominales (One-Hot)
categorical_cols = [
    "host_identity_verified",
   "instant_bookable",
    "cancellation_policy",
   "room type"
]
#data = pd.get_dummies(data, columns=categorical_cols, drop_first=True)
```

```python
categorical_cols = [
    "host_identity_verified",
    "instant_bookable",
    "cancellation_policy",
    "room type"
]

encoder = OneHotEncoder(drop="first", sparse_output=False)

encoded = encoder.fit_transform(data[categorical_cols])

encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(categorical_cols)
)

encoded_df.index = data.index

# Drop old columns
data = data.drop(columns=categorical_cols)

# Concat safely
data = pd.concat([data, encoded_df], axis=1)
```
